```python
def custom_resampler(arr):
    if len(arr) > 0:
        return {'closeAsk': arr.iloc[-1,0],
                'closeBid': arr.iloc[-1,1],
                'complete': True, 
                'highAsk': max(arr.iloc[:,3]),
                'highBid': max(arr.iloc[:,4]), 
                'lowAsk': min(arr.iloc[:,5]), 
                'lowBid': min(arr.iloc[:,6]), 
                'openAsk': arr.iloc[0,7], 
                'openBid': arr.iloc[0,8], 
                'volume': sum(arr.iloc[:,9])}
    else:
        return {'closeAsk': float("NaN"), 
                'closeBid': float("NaN"),
                'complete': True, 
                'highAsk': float("NaN"),
                'highBid': float("NaN"), 
                'lowAsk': float("NaN"), 
                'lowBid': float("NaN"), 
                'openAsk': float("NaN"), 
                'openBid': float("NaN"), 
                'volume': 0}
```

```python
data.resample('15T', how=custom_resampler)
```
